```python
# 导入相应的包
import networkx as nx 
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import *
import numpy as np

# 定义自定义字体，文件名从1.b查看系统中文字体中来 
myfont = FontProperties(fname='/Library/fonts/Songti.ttc')   
# 解决负号'-'显示为方块的问题   
matplotlib.rcParams['axes.unicode_minus']=False

#控制numpy输出小数位数
np.set_printoptions(precision=3)
```

```python
#定义图的节点和边 
nodes=['0','1','2','3','4','5','a','b','c'] 
edges=[('0','0',1),('0','1',1),('0','5',1),('0','5',2),('1','2',3),('1','4',5),('2','1',7),('2','4',6),('a','b',0.5),('b','c',0.5),('c','a',0.5)]
```

```python
#定义一个无向图和有向图 
G1 = nx.Graph() 
G1.add_nodes_from(nodes) 
G1.add_weighted_edges_from(edges) 
 
G2 = nx.DiGraph() 
G2.add_nodes_from(nodes) 
G2.add_weighted_edges_from(edges) 
 
pos1=nx.circular_layout(G1) 
pos2=nx.circular_layout(G2) 
```

```python
#画出无向图和有向图 
plt.subplots(1,2,figsize=(15,6)) 
plt.subplot(121) 
nx.draw(G1,pos1, with_labels=True, node_size=1000, font_size=20, font_weight='bold', width=3, alpha=0.6) 
plt.title('无向图',fontproperties=myfont, fontsize=20) 
plt.axis('on') 
plt.xticks([]) 
plt.yticks([]) 
 
plt.subplot(122) 
nx.draw(G2,pos2, with_labels=True, node_size=1000, font_size=20, font_weight='bold', width=3, alpha=0.6) 
plt.title('有向图',fontproperties=myfont, fontsize=20) 
plt.axis('on') 
plt.xticks([]) 
plt.yticks([]) 
 
plt.show() 
```

```python
#邻接矩阵 
A = nx.adjacency_matrix(G1) 
print('邻接矩阵:\n',A.todense())
```

```python
#关联矩阵 
I = nx.incidence_matrix(G1) 
print('\n关联矩阵:\n',I.todense())
```

```python
#拉普拉斯矩阵 L=D-A
L=nx.laplacian_matrix(G1) 
print('\n拉普拉斯矩阵:\n',L.todense()) 
```

```python
#标准化的拉普拉斯矩阵 
NL=nx.normalized_laplacian_matrix(G1) 
print('\n标准化的拉普拉斯矩阵:\n',NL.todense()) 
```

```python
#有向图拉普拉斯矩阵 
DL=nx.directed_laplacian_matrix(G2) 
print('\n有向拉普拉斯矩阵:\n',DL) 
```

```python
#拉普拉斯算子的特征值 
LS=nx.laplacian_spectrum(G1) 
print('\n拉普拉斯算子的特征值:\n',LS)
```

```python
#邻接矩阵的特征值 
AS=nx.adjacency_spectrum(G1) 
print('\n邻接矩阵的特征值:\n',AS)
```

```python
#无向图的代数连通性 
AC=nx.algebraic_connectivity(G1) 
print('\n无向图的代数连通性:\n',AC)
```

```python
#图的光谱排序 
SO=nx.spectral_ordering(G1) 
print('\n图的光谱排序:\n',SO) 
```
